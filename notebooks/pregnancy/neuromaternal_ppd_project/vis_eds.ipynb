```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from polpo.preprocessing.load.pregnancy import (
    NeuroMaternalTabularDataLoader,
)
```

```python
pipe = NeuroMaternalTabularDataLoader(
    keep_mothers=True,
    keep_control=False,
)

df = pipe()

df
```

```python
val = 10.0
var = "EDS.Total"

counts = [(df[var] < val).sum(), (df[var] >= val).sum()]

plt.bar(["< {}".format(val), "â‰¥ {}".format(val)], counts)
plt.xlabel("EDS")
plt.ylabel("Count")

plt.show()
```

```python
ax = df[var].hist(grid=False)

ax.set_xlabel("EDS")
ax.set_ylabel("Count")

ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
```
